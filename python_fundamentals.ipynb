# Python Fundamentals

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load spreadsheets into dataframes
articles = pd.read_csv("articleInfo.csv")
authors = pd.read_csv("authorinfo.csv")
```

## Merge “articleInfo.csv” and “authorInfo.csv” into one data frame based on “Article No.”, fill all empty cells with the value of 0, and answer the following questions.

```python
# Merge articles onto authors where "Article No." in authors equals "Articles No."
publications = pd.merge(authors, articles, how = "left", on = "Article No.")
publications.head(10)
```

```python
# Place 0's in the publications dataframe where an empty value was found
publications.fillna(0, inplace = True)
publications.head(10)
```

## 1. Plot the yearly_publication figure, in which the x-axis is the year, the y-axis is the number of articles published during that year.

```python
# Drop duplicate articles to avoid counting an article multiple times
unique_articles = publications.drop_duplicates(subset=['Article No.'])

# Publictations per year
pub_years = unique_articles["Year"].value_counts()
pub_years
```

```python
# Yearly_publication plot
plt.figure(figsize=(16, 9))
sns.lineplot(x = pub_years.index, y = pub_years)
plt.title("yearly_publication")
plt.show()
```

## 2. Plot the yearly_citation figure, in which the x-axis is the year, the y-axis is the total number of citations during that year.

```python
# Citations per year
cit_years = unique_articles.groupby("Year").Citation.sum()
cit_years
```

```python
# Yearly_Citation plot
plt.figure(figsize=(16, 9))
sns.lineplot(x = cit_years.index, y = cit_years)
plt.title("yearly_citation")
plt.show()
```

## 3. Plot the figure of the number of publications across countries. You may use any available python libraries, such as pygal_maps_world, geopandas, or others.

```python
# Some data had to be cleaned up. Some misspelling and spaces added had to be fixed

# Publications per Country: Avoids repeated articles
pub_country = publications.groupby('Country', sort= True)["Article No."].nunique().to_frame('count')
# Drop Empty Value
pub_country = pub_country.drop(index = 0)
# Drop "Bristol", not a country
pub_country = pub_country.drop(index = "Bristol")

pub_country.reset_index(inplace = True)

# List of country abbreviations to work with pygal_maps_world, Korea is assumed to be the Republic of Korea (kr)
abbreviations = ['au', 'ca', 'cl', 'cn', 'cy', 'cz', 'dk', 'fr', 'de',
                 'gr', 'hk', 'in', 'ie', 'il', 'it', 'kr', 'kg', 'li', 
                 'my', 'mx', 'nz', 'no', 'pk', 'ps', 'ru', 'sk', 'za', 
                 'es', 'ch', 'tw', 'us', 'ua', 'ae', 'gb']
pub_country["Country"] = abbreviations
# Transform into a dictionary to be used in pygal
dic = pub_country.set_index("Country").T.to_dict('records')[0]
dic
```

```python
import pygal  as pyg
worldmap_chart = pyg.maps.world.World()
worldmap_chart.title = 'Publications Across Countries'
worldmap_chart.add('Publications', dic)
worldmap_chart.render_to_file('abc.svg')
```

## 4. What are the top 5 institutions that have the most published articles in this area?

```python
# Publications per Institution
pub_institution = publications.groupby('Author Affiliation', sort= True)["Article No."].nunique().to_frame('count')
# Sort by count
pub_institution = pub_institution.sort_values(by = ['count'], ascending = False)
# Display the first five entries. Sorted so these are the top 5 institutions
pub_institution.head(5)
```
